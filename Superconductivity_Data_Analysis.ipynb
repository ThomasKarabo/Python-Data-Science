```python
#Given the data of the sample
m=1.58e-5 #The mass in units g
ρ=8.57 #The density in units g/cm^3
```

```python
V = m/ρ
V
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
FC_path = "/content/FC.csv"
ZFC_path = "/content/ZFC-FW.csv"
```

```python
FC_data = pd.read_csv(FC_path)
```

```python
FC_data
```

```python
FC_data = FC_data.dropna(axis=1)
```

```python
FC_data
```

```python
FC_data['X(T) FC'] = FC_data['DC Moment (emu)'] / (FC_data['Magnetic Field (Oe)'] * V)
```

```python
FC_data
```

```python
ZFC_data = pd.read_csv(ZFC_path)
```

```python
ZFC_data = ZFC_data.dropna(axis=1)
```

```python
ZFC_data
```

```python
ZFC_data['X(T) ZFC'] = ZFC_data['DC Moment (emu)'] / (ZFC_data['Magnetic Field (Oe)'] * V)
```

```python
ZFC_data
```

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(FC_data['Temperature (K)'],FC_data['X(T) FC'] , label='FC', color='blue', linestyle='-')
ax.plot(ZFC_data['Temperature (K)'],ZFC_data['X(T) ZFC'] , label='ZFC-FW', color='red', linestyle='-')

# Add labels and a title
ax.set_xlabel('Temperature (K)')
ax.set_ylabel('X(T)')
ax.set_title('Temperature Vs. X(T)')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
